Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular data visualization library in Python that allows users to create a wide range of static, animated, and interactive visualizations. It provides a variety of tools for creating plots, graphs, and other visual representations of data.

Matplotlib is widely used in data analysis, scientific research, and various fields where visualizing data is crucial.

    1. Line Plot: Displays data points connected by straight lines.
    2. Scatter Plot: Represents individual data points as dots on a two-dimensional plane.
    3. Bar Plot: Uses rectangular bars to represent data values, typically for categorical data.
    4. Histogram: Displays the distribution of a continuous variable by dividing it into bins.
    5. Pie Chart: Represents data as slices of a circular pie, suitable for showing proportions.

Q2: What is a scatter plot?

A scatter plot is a type of data visualization that uses individual data points to represent values for two different variables on a two-dimensional plane. Each data point is represented as a dot, and the position of the dot on the plot corresponds to the values of the two variables. 

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter Plot")
plt.show()
# Note: Also add title, xlabel, and ylabel to the plot.
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

The subplot() function in Matplotlib is used to create multiple subplots within a single figure.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Data for line plots
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Line 4")

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A bar plot, also known as a bar chart or bar graph, is a data visualization that uses rectangular bars to represent data values. It is particularly useful for displaying and comparing categorical data or discrete values.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
color = np.array(["red", "blue", "black", "yellow"])
plt.bar(company,profit , color = color)
plt.xlabel("company")
plt.ylabel("profit")
plt.title("BAR GRAPH")
plt.show()
```

```python
plt.barh(company,profit , color = color)
plt.xlabel("company")
plt.ylabel("profit")
plt.title(" HORIZONTAL BAR GRAPH")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a statistical data visualization that displays the distribution of a dataset's values. It provides a visual summary of the minimum, first quartile (25th percentile), median (second quartile or 50th percentile), third quartile (75th percentile), and maximum values of the dataset. Additionally, box plots can help identify outliers in the data.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data, labels=["Box 1", "Box 2"])
plt.title("Box Plot")
plt.ylabel("Values")
plt.show()
```
